# Distribuição de probabilidade

```python
import matplotlib.pyplot as fig
import numpy as np
```

### Função random

```python
aleat = [np.random.randint(-3,3) for i in range(100)] #para numeros continuos utilizar random.ramdom(a,b)
fig.plot(aleat)
fig.show()
```

### Distribuição uniforme

#### Definição: Se uma v.a. discreta X assume valores x1,x2,...,xk com igual probabilidade, dizemos que X segue uma distribuição uniforme discreta cuja fp é dada por $\frac{1}{k} , x \in (x1,x2,...,xk)$ . Na formula random.uniform(a,b) temos que  $\frac{1}{b-a}$ se  $a\leq x\leq b$

Vamos gerar 100 numeros aleatorio entre -1 e 1

```python
#aleat = [np.random.uniform(-1,1) for i in range(100)]
aleat = np.random.uniform(-1,1,100)
#print(aleat)

fig.subplot(211) #2 linhas 1 coluna 1 grafico
fig.plot(aleat, color = 'black')
fig.xlabel('eixo horizontal')
fig.ylabel('numeros aleatorios entre -1 e 1')

fig.subplot(212) #2 linhas 1 coluna 2 grafico
fig.hist(aleat, bins = 10) #10 classes
fig.xlabel('classes')
fig.ylabel('frequência')
fig.show()
```

Foram criadas 10 classes, ou seja, foi percorrido o vetor aleat e contado o número de valores entre -1 e -0.75, por exemplo

### Distribuição Binomial

```python
#aleat = [np.random.binomial(5,0.1) for i in range(100)]
aleat = np.random.binomial(500,0.1,100) #(number of trials, probability of trial, number of test)
print(aleat)

fig.subplot(211) #2 linhas 1 coluna 1 grafico
fig.plot(aleat, color = 'black')
fig.xlabel('eixo horizontal')
fig.ylabel('numeros aleatorios entre')

fig.subplot(212) #2 linhas 1 coluna 2 grafico
fig.hist(aleat, bins = 10) #10 classes
fig.xlabel('classes')
fig.ylabel('frequência')
fig.show()
```

Exemplos

Resultado de jogar uma moeda 10 vezes, testado 100 vezes
cada valor encontrado é quantos sucessos eu obtive, sendo que 0 <= sucesso <= n, aumente a probabilidade para 1 e veja o que ocorre

```python
n, p = 10, 0.3  # number of trials, probability of each trial
s = np.random.binomial(n, p, 100)
s
```

Um exemplo do mundo real. Uma empresa perfura 9 poços de exploração de petróleo, cada um com uma probabilidade de sucesso estimada de 0,1. Todos os nove poços falham. Qual é a probabilidade de isso acontecer?

```python
#Vamos fazer 20.000 tentativas do modelo e contar o número que gera resultados positivos zero.
sum(np.random.binomial(9, 0.1, 20000) == 0)/20000
```

### Distribuição Geometrica

```python
aleat = np.random.geometric(0.2,100)
print(aleat)

fig.subplot(211) #2 linhas 1 coluna 1 grafico
fig.plot(aleat, color = 'black')
fig.xlabel('eixo horizontal')
fig.ylabel('numeros aleatorios entre')

fig.subplot(212) #2 linhas 1 coluna 2 grafico
fig.hist(aleat, bins = 10) #10 classes
fig.xlabel('classes')
fig.ylabel('frequência')
fig.show()
```

Exemplos

```python
#jogar uma moeda com viciada com probabilidade de sucesso de 0,7. Realizar o experimento 30 vezes
z = np.random.geometric(p=0.7, size=10)
z
```

cada valor do vetor é quantas vezes foi preciso pra alcançar sucesso

```python
#Draw ten thousand values from the geometric distribution, with the probability of an individual success equal to 0.35:
z = np.random.geometric(p=0.35, size=10000)
```

How many trials succeeded after a single run?

```python
(z == 1).sum() / 10000.
```

```python
z = np.random.geometric(p=0.2, size=1000)
np.mean(z)
```
